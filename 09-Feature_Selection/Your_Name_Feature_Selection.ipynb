# Predicting Annual Healthcare Costs: A Feature Selection Case Study

## The Business Question

You're an analyst at a health insurance company. Leadership wants a model that
predicts a new customer's **annual healthcare cost** from intake survey data,
so pricing can be set fairly and consistently.

Before anyone touches a model, your manager asks the team: *"Which factors do
you think matter most?"*

Almost everyone in the room gives the same answer: **age** and **BMI**. It's
the intuitive answer -- we all assume health costs climb steadily as people
get older and heavier.

Our job is to let the data answer the question instead of intuition. We'll walk through data prep, EDA, feature engineering, and -- the focus of this lecture -- **feature selection**, using filter, wrapper, and embedded methods. Keep an eye out: by the end, one feature is going to completely upend what the team assumed.

## Ronald A Fisher

In the late 1950s, Sir Ronald Fisher — one of the founders of modern statistics — publicly questioned the by-then widely accepted causal interpretation of the link between smoking and lung cancer, and his forcefully stated views set off a controversy that has continued to resurface ever since.

**His actual argument** wasn't "smoking is harmless" — it was a methodological one, rooted in "correlation isn't causation." He proposed that the smoking–cancer association could instead reflect either reverse causation (an early, undetected lung condition causing people to smoke) or a common underlying cause — in his framing, a genetic predisposition that made someone both more likely to smoke and more likely to develop cancer. That became known as his "constitutional hypothesis," and it became a core part of the tobacco industry's broader strategy of manufacturing scientific-sounding doubt around the health effects of cigarettes.

Fisher wasn't a crank — he was a smoker himself, and by most accounts a genuinely rigorous scientist elsewhere in his career. A retrospective review in the *American Journal of Epidemiology* put it bluntly: although Fisher made invaluable contributions to statistics, his analysis of the causal link between lung cancer and smoking was flawed by an unwillingness to examine the full body of evidence and by drawing conclusions prematurely — and his views may have been shaped by personal and professional conflicts, his tobacco-industry consulting work, and the fact that he smoked himself.

At that time, he was serving as a paid consultant to the Tobacco Manufacturers' Standing Committee, and the tobacco industry made active use of his reputation and outspoken support for a genetic explanation linking smoking behavior and cancer risk.

A founding figure of hypothesis testing arguing "don't confuse correlation with causation" in a case where the causal story turned out to be true, partly because of an undisclosed conflict of interest.

### 4b. Wrapper Methods

Wrapper methods evaluate feature subsets by using an actual machine learning model to test different combinations and score their performance. Unlike filter methods, they treat feature selection as a search problem, adding or removing features to see which specific combination yields the highest model accuracy.

Unlike the filter methods, RFE actually fits a model and removes features
based on their real contribution to it, one at a time. It converges on
`age`, `bmi`, `weight_kg`, `height_cm`, and `smoker` -- notice it kept the
whole BMI cluster rather than resolving the redundancy itself. That's a good
teaching point: **RFE optimizes predictive contribution, not multicollinearity**
-- it's not a substitute for the VIF check we already ran, it's a complement
to it.


### Other Wrapper Methods

- **Forward Selection** — starts with zero features, then adds one at a time — whichever improves the model most at each step — until adding more no longer helps.

- **Backward Elimination** — the opposite: starts with *all* features, then removes the least useful one each round until only the best remain. (RFE is essentially a formalized version of this.)

- **Stepwise Selection (bidirectional)** — combines both directions: at each step it can add a promising feature *or* remove one that's stopped pulling its weight, checking after every change.

- **RFECV (Recursive Feature Elimination with Cross-Validation)** — same idea as RFE, but instead of you picking a fixed number of features to keep, it uses cross-validation to automatically find the number of features that gives the best generalization performance.

- **Exhaustive Feature Selection** — tries *every possible combination* of features and keeps whichever subset performs best. Guarantees the optimal subset for your chosen model, but the number of combinations explodes fast, so it's really only practical with a small number of candidate features.

- **Sequential Feature Selector (from `mlxtend`)** — a more flexible library implementation that lets you do forward or backward selection with your choice of model and scoring metric, plus built-in cross-validation.

The common thread with all of these is that they all repeatedly **fit an actual model** and judge features by how much they help that specific model's performance — that's what makes them "wrapper" methods, as opposed to filter methods (which never touch a model) or embedded methods (where selection happens as a byproduct of training, like Lasso zeroing out coefficients).

### 4c. Embedded Methods

Unlike filter methods (which screen data before training) or wrapper methods (which run separate search loops), embedded methods select the best features automatically while the model learns its parameters.
